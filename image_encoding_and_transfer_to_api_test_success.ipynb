```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
import base64
```

```python
with open('bugs.jpg', 'rb') as f:
    b64_img_string = base64.b64encode(f.read()).decode('utf-8')
```

```python
b64_img_string
```

```python
type(b64_img_string)
```

```python
isinstance(b64_img_string, bytearray), isinstance(b64_img_string, bytes)
```

```python
image = np.fromstring(base64.b64decode(b64_img_string), dtype=np.uint8)
img = cv2.imdecode(image, cv2.IMREAD_COLOR)
```

```python
image
```

```python
plt.imshow(img)
```
